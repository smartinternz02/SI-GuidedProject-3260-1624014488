```python
weather_data['datetime_utc'] = pd.to_datetime(weather_data['datetime_utc'])
```

```python
weather_data.set_index('datetime_utc',inplace=True)
weather_data
```

```python
weather_data = weather_data.resample('D').mean()
```

```python
weather_data
```

```python
weather_data = weather_data[[' _tempm']]
```

```python
weather_data.info()
```

```python
weather_data.reset_index(inplace=True)
```

```python
weather_data.reset_index(inplace=True)
```

```python
weather_data.rename(columns = {'datetime_utc':'ds',' _tempm':'y'},inplace=True)
```

```python
weather_data.head()
```

```python
weather_data["ds"]
```

```python
plt.plot(weather_data["ds"],weather_data["y"],ms=20,mec ="r")
plt.xlabel("date")
plt.ylabel("Temperature")
plt.title("weather prediction")
plt.grid()
```

```python
plt.figure(figsize=(12,8))
plt.plot(weather_data.set_index(["ds"]))
```

```python
model = Prophet()
```

```python
model.fit(weather_data)
```

```python
future = model.make_future_dataframe(periods=365,freq="D")
future.tail()
```

```python
forecast = model.predict(future)
```

```python
forecast.head()
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].head()
```

```python
model.plot(forecast)
```

```python
fig = model.plot_components(forecast)
```

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(model,initial='730 days',period='180 days',horizon='365 days')
df_cv.head()
```

```python
from fbprophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(df_cv,metric='rmse')
```
